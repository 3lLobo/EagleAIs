```python
from ultralytics import YOLO

model_path = "/home/wolf/worqspace/EagleEyez/runs/detect/train/weights/best.pt"
model = YOLO(model_path)  # load a pre-trained model
```

```python
import os
from PIL import Image

img_dir = lambda x: f'/home/wolf/worqspace/EagleEyez/data/png/s{x}/'
images = []
for n in [1,2]:
  #  Loop over all the images
  for img in os.listdir(img_dir(n)):
    # Append the full path of the image to the list
    if img.endswith('.png'):
      image = Image.open(img_dir(n) + img)
      images.append(image)
```

```python

model.predict(images, save=True)
```
